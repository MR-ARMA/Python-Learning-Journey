```python
class Book:
    book_count = 0
    def __init__(self, title, author, pages):
        self.title = title
        self.pages = pages
        self.author = author
        Book.book_count += 1
    
    def description(self):
        return f'Title: {self.title}, Author: {self.author}, Pages: {self.pages}'
    
    def is_long(self):
        return self.pages > 500
```

**inheriting**

```python
class Vehicle:
    def __init__(self, model, year):
        self.model = model
        self.year = year
    
    def dispaly_info(self):
        return f'Model: {self.model}, Year: {self.year}.'
    
```

```python
class Car(Vehicle):
    pass
```

```python
car1 = Car('Benz', 2025)
```

```python
car1.dispaly_info()
```

```python
class Car(Vehicle):
    def __init__(self, model, year, trunk_capacity):
        super().__init__(model, year)  # Calls the __init__ of Vehicle
        self.trunk_capacity = trunk_capacity  # New attribute specific to Car

    def display_car_info(self):  # New method specific to Car
        print(f"{self.year} {self.model} with trunk capacity: {self.trunk_capacity} liters")

    # Overriding an inherited method
    def display_info(self):
        print(f"This is a {self.year} {self.model}.")
```

```python
my_car = Car("Civic", 2020, 450)
```

```python
my_car.display_car_info()
```

```python
my_car.display_info()
```
